```python
#Câu 3
import pandas as pd

# 1. Tải dữ liệu lên dataframe
df = pd.read_csv('dulieuxettuyendaihoc.csv')

# 2. In 10 dòng đầu tiên
print("--- 10 DÒNG ĐẦU TIÊN ---")
display(df.head(10))

# 3. In 10 dòng cuối cùng
print("\n--- 10 DÒNG CUỐI CÙNG ---")
display(df.tail(10))
```

```python
#Câu 4
print("--- Thống kê cột dân tộc trước khi xử lý ---")
tan_so_dt = df['DT'].value_counts(dropna=False)
tan_suat_dt = df['DT'].value_counts(dropna=False, normalize=True) * 100

thong_ke_dt = pd.DataFrame({'Tần số': tan_so_dt, 'Tần suất (%)': tan_suat_dt})
display(thong_ke_dt)

# 2. Thay thế dữ liệu thiếu bằng số 0
df['DT'] = df['DT'].fillna(0)
print("\n-> Đã xử lý điền giá trị 0 cho các vị trí thiếu ở cột 'DT'.")
```

```python
# CÂU 5: XỬ LÝ DỮ LIỆU THIẾU CHO BIẾN T1

# 1. Thống kê dữ liệu thiếu trước khi xử lý
print("Số lượng dữ liệu thiếu ở biến T1:", df['T1'].isnull().sum())

# 2. Tính giá trị trung bình (Mean) của cột T1
mean_t1 = df['T1'].mean()
print(f"Giá trị Mean của T1 dùng để điền: {mean_t1}")

# 3. Thay thế dữ liệu thiếu bằng Mean
df['T1'] = df['T1'].fillna(mean_t1)
print("-> Đã xử lý thiếu cho cột 'T1'.")
```

```python
# CÂU 6: XỬ LÝ DỮ LIỆU THIẾU CHO CÁC BIẾN ĐIỂM CÒN LẠI BẰNG MEAN

# Tự động tìm các cột điểm số và điền giá trị mean của chính cột đó nếu bị thiếu
cac_cot_so = df.select_dtypes(include=['float64', 'int64']).columns

for col in cac_cot_so:
    if df[col].isnull().sum() > 0:
        df[col] = df[col].fillna(df[col].mean())

print("-> Đã xử lý toàn bộ dữ liệu thiếu cho các biến điểm số bằng phương pháp Mean.")
```

```python
# CÂU 7: TÍNH ĐIỂM TRUNG BÌNH MÔN TBM1, TBM2, TBM3

# Áp dụng công thức theo yêu cầu đề bài
df['TBM1'] = (df['T1']*2 + df['L1'] + df['H1'] + df['S1'] + df['V1']*2 + df['X1'] + df['D1'] + df['N1']) / 10
df['TBM2'] = (df['T2']*2 + df['L2'] + df['H2'] + df['S2'] + df['V2']*2 + df['X2'] + df['D2'] + df['N2']) / 10
df['TBM3'] = (df['T3']*2 + df['L3'] + df['H3'] + df['S3'] + df['V3']*2 + df['X3'] + df['D3'] + df['N3']) / 10

# Hiển thị thử 5 dòng đầu của kết quả vừa tính
display(df[['TBM1', 'TBM2', 'TBM3']].head())
```

```python
# CÂU 8: XẾP LOẠI HỌC LỰC XL1, XL2, XL3

# Định nghĩa hàm quy đổi điểm sang ký hiệu xếp loại
def xep_loai(diem):
    if diem < 5.0:
        return 'Y'
    elif diem < 6.5:
        return 'TB'
    elif diem < 8.0:
        return 'K'
    elif diem < 9.0:
        return 'G'
    else:
        return 'XS'

# Áp dụng hàm xếp loại cho từng năm học
df['XL1'] = df['TBM1'].apply(xep_loai)
df['XL2'] = df['TBM2'].apply(xep_loai)
df['XL3'] = df['TBM3'].apply(xep_loai)

# Hiển thị kết quả điểm và xếp loại để đối chiếu
display(df[['TBM1', 'XL1', 'TBM2', 'XL2', 'TBM3', 'XL3']].head())
```

```python
# CÂU 9: CHUYỂN ĐỔI SANG THANG ĐIỂM 4 CỦA MỸ (MIN-MAX NORMALIZATION)

# Chuyển đổi từ thang điểm 10 sang thang điểm 4
df['US_TBM1'] = (df['TBM1'] / 10) * 4
df['US_TBM2'] = (df['TBM2'] / 10) * 4
df['US_TBM3'] = (df['TBM3'] / 10) * 4

# Hiển thị kết quả điểm hệ Việt Nam và hệ Mỹ để kiểm tra
display(df[['TBM1', 'US_TBM1', 'TBM2', 'US_TBM2', 'TBM3', 'US_TBM3']].head())
```

```python
# CÂU 10: TẠO BIẾN KẾT QUẢ XÉT TUYỂN (KQXT)

# Nếu TBM3 >= 5.0 thì KQXT = 1 (Đạt), ngược lại KQXT = 0 (Không đạt)
df['KQXT'] = df['TBM3'].apply(lambda x: 1 if x >= 5.0 else 0)

# Thống kê số lượng thí sinh Đạt và Không đạt
print("--- Thống kê kết quả xét tuyển (1: Đạt, 0: Trượt) ---")
print(df['KQXT'].value_counts())

# Hiển thị thử vài dòng kết quả
display(df[['TBM3', 'KQXT']].head())
```

```python
# CÂU 11: LƯU TRỮ VÀ XUẤT DỮ LIỆU ĐÃ XỬ LÝ

# Xuất DataFrame hoàn chỉnh ra file CSV theo đúng tên yêu cầu của đề bài
df.to_csv('processed_dulieuxettuyendaihoc.csv', index=False)

print("-> Đã thực hiện câu 11: Lưu và xuất file 'processed_dulieuxettuyendaihoc.csv' thành công!")
print("Số lượng dòng và cột hiện tại của file:", df.shape)
```
